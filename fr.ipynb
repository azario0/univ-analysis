```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Tracé du graphique

```python
# Définir la taille de la figure
plt.figure(figsize=(10, 8))
# Tracé de la heatmap de corrélation avec les valeurs
dataplot = sns.heatmap(df.corr(numeric_only=True), cmap="YlGnBu", 
                       annot=True, fmt='.2f')
plt.yticks(fontsize=12)
# Afficher la heatmap
plt.show()
```
